```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from  sklearn.impute import SimpleImputer, MissingIndicator
```

```python
df = pd.read_csv('/content/train.csv')
df.head()
```

```python
df['site_eui'].isnull().sum()
```

```python
df['site_eui'] = np.where(df['site_eui'].isnull(), 1, 0)
```

```python
X = df[['floor_area', 'energy_star_rating', 'max_wind_speed']]
y = df['site_eui']
```

```python
df['site_eui'].isnull().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 22)
```

```python
X_train_original = X_train.copy()
X_test_original = X_test.copy()
```

```python
X_train_original.head()
```

```python
X_test_original
```

```python
#missing indicator using for loop, missing:0 and not-missing:1
for var in ['floor_area', 'energy_star_rating', 'max_wind_speed']:
    X_train[var + '_missing'] = np.where(X_train[var].isnull(), 1, 0) #missing 0, not missing 1
    X_test[var + '_missing'] = np.where(X_test[var].isnull(), 1, 0)
```

```python
X_train['energy_star_rating'].isnull().sum()
```

```python
X_train['energy_star_rating_missing'].sum()
```

```python
#sklearn add_indicator
imputer = SimpleImputer(strategy="mean",add_indicator=True)
X_train_original_transformed = imputer.fit_transform(X_train_original)
X_test_original_transformed = imputer.transform(X_test_original)
```

```python
X_test_original_transformed
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
reg = LinearRegression()
reg.fit(X_train_original_transformed, y_train)
```

```python
y_pred = reg.predict(X_test_original_transformed)
```

```python
r2_score(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```
