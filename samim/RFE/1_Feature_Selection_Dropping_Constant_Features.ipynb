### Feature Selection- Dropping constant features
In this step we will be removing the features which have constant features which are actually not important
for solving the problem statement

#**Definition:**
###Constant features are variables that have only one unique value throughout the entire dataset. Since they don't contribute any variability, they are usually not informative for modeling tasks.

#**Reasons to Drop Constant Features:**

###**No Discriminatory Power:** Constant features do not help the model differentiate between observations, as they have the same value across all instances.
###**Computational Efficiency:** Removing constant features reduces the computational load and memory requirements for training and predicting with machine learning models.
#**Procedure:**

###**Identify Constant Features:** Use statistical measures or simple checks to identify features with constant values. Common methods include using variance thresholds or directly checking for unique values.

###**Drop Constant Features:** Once identified, these features can be dropped from the dataset.

```python
# Import pandas to create DataFrame
import pandas as pd

# Make DataFrame of the given data
data = pd.DataFrame({"A":[1,2,4,1,2,4],
                    "B":[4,5,6,7,8,9],
                    "C":[0,0,0,0,0,0],
                    "D":[1,1,1,1,1,1]})
```

```python
data.head()
```

##### Variance Threshold
Feature selector that removes all low-variance features.

This feature selection algorithm looks only at the features (X), not the desired outputs (y), and can thus be used for unsupervised learning.

```python
### It will zero variance features
from sklearn.feature_selection import VarianceThreshold
var_thres=VarianceThreshold(threshold=0)
var_thres.fit(data)
```

```python
var_thres.get_support()
```

```python
data.columns[var_thres.get_support()]
```

```python
constant_columns = [column for column in data.columns
                    if column not in data.columns[var_thres.get_support()]]

print(len(constant_columns))
```

```python
for feature in constant_columns:
     print(feature)
```

```python
data.drop(constant_columns,axis=1)
```

#**Method2**


```python
from sklearn.feature_selection import VarianceThreshold
import pandas as pd

# Make DataFrame of the given data
df = pd.DataFrame({"A":[1,2,4,1,2,4],
                    "B":[4,5,6,7,8,9],
                    "C":[0,0,0,0,0,0],
                    "D":[1,1,1,1,1,1]})
print(df.head())
X = df

# Assuming 'X' is your feature matrix
# Set a threshold for variance (e.g., 0 for constant features)
threshold = 0

# Use VarianceThreshold to remove features below the threshold
selector = VarianceThreshold(threshold=threshold)
X_selected = selector.fit_transform(X)
print(X_selected)

# Get the indices of selected features
selected_indices = selector.get_support(indices=True)
print(selected_indices)

# Create a DataFrame with the selected features
X_selected_df = pd.DataFrame(X_selected, columns=X.columns[selected_indices])
print(X_selected_df)
```
